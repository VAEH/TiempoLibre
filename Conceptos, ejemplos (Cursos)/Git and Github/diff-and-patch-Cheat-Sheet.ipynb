## Diffing Files (Diferenciar archivos)
**`diff`** Se utiliza cuando queremos comparar 2 tipos de achivo de código y encontrar alguna diferencia dentro de una función y/o alguna parte del código.
Cuando se utiliza, obtenemos solo las lineas que son diferentes entre dos archivos. Es mucho más facil encontrar la diferencia cuando solo tenemos dos lineas.



¿Ves los símbolos al principio de cada una de esas líneas? El **símbolo menos** que nos dice que la primera línea fue eliminada del primer archivo, y el **símbolo mayor** que nos dice que la segunda línea fue agregada al segundo archivo. En otras palabras, la línea antigua fue reemplazada por la nueva. En este ejemplo, teníamos una línea que fue reemplazada por una nueva. Este es un cambio común al modificar código, pero no la única posibilidad.

<img src="Images/seccion-diff.png"  style="width: 400px;"/>
Podemos ver que `diff` divide los cambios en dos secciones separadas. La sección que comienza con **5c5,6** muestra una línea en el primer archivo que fue reemplazada por dos líneas diferentes en el segundo archivo. El número al principio de esta sección indica el número de línea en el primer y segundo archivo. **La c** entre los números significa que una línea fue cambiada.

La sección que comienza con 11a13,15 muestra tres líneas que son nuevas en el segundo archivo. La ***a*** significa que agrego. Aunque las lineas que agrego son algo extrañas referente a dicho orden. 
Para entender que paso, se usa la bandera `dash -u ` para que `diff` muestre las diferencias en otro formato.
<img src="Images/nuevos-valores.png"  style="width: 400px;"/>
Este formato unificado muestra las lineas de cambio junto con algún contexto. usando el signo menos para marcar las líneas que se eliminaron, y el signo más para marcar las líneas que se agregaron. El contexto extra nos permite saber mejor qué está pasando con el cambio que estamos dificando. 

 Podemos ver que el nuevo archivo en realidad tiene un bloque completamente nuevo si que es parte de una cadena de condicionales que se ve muy similar, y es por eso que con la salida diff que vimos antes, fue un poco confuso qué líneas se habían agregado.

* `wdiff` Resalta las palabras que han cambiado en una archivo en lugar de trabajar linea por linea como lo hace `diff`


### Applying Changes
Cuando se quiere dejar claro un cambio dentro del código, se puede enviar aun compañero un `diff` con el cambio para que puedan ver cómo se ve el código modificado, para eso usamos la siguiente linea:
<br>
> diff-u old_file new_ file > change.diff 

El **signo mayor** redirige la salida del comando diff a un archivo. Con este comando estamos generando un archivo llamado *change.diff* con el contenido del comando `diff -i `. usando la bandera `-u`, incluimos más context que ayuda a la persona que esta leyendo el archivo a entender lo que está pasando con el cambio. 

El **archivo diff** o **archivo patch**, incluye todos los cambios entre el archivo antiguo y el nuevo, más el contexto adicional necesario para entender los cambios y aplicar esos cambios de nuevo al archivo original. 

**`patch`**, toma un archivo generado por `diff` y aplica los cambios al archivo original.



En la primer parte, tenemos el código que comprueba el porcentaje de la CPU que esta actualmente en uso y da unas alertas cuando pasa del umbral de 75%, dicho script se comparte con algunos colegas y encuentran problemas.

La segunda parte muestra la diferencia de un compañero con la solución de nuestro problema, evidenciamos que nuestro colega hizo dos cambios. Así que tenemos el archivo diff y queremos aplicarlo a nuestro script.<br>
Vamos a utilizar el simbolo `<` para redirigir el contenido del archivo a entrada estandar.

Así que le dijimos a parche que aplicara los cambios que vienen de cpu_usage.diff a nuestro archivo cpu_usage.py. Obtenemos una sola línea que dice que el archivo fue parcheado, lo que significa que hemos aplicado correctamente los cambios.

*  Mientras que diff es el comando que genera la diferencia entre dos archivos, patch es el comando que aplica esas diferencias al archivo original.



### Aplicación Practica diff y patch
Imagínese esto, un colega nos está pidiendo ayuda para arreglar un script llamado `disk_usage.py`. El objetivo de la secuencia de comandos es comprobar cuánto espacio en disco se utiliza actualmente, y imprime un error si es demasiado poco espacio para el funcionamiento normal. Pero el script está actualmente roto porque tiene algunos errores. Ayudaremos a nuestro colega a corregir esos errores para demostrar cómo usar `diff` y `patch`.<br> Antes de cambiar nada, hagamos un par de copias del guión. Agregaremos `_original` a una copia, que conservaremos sin modificar y usaremos para la comparación y `_fixed` a la otra copia, que usaremos para reparar nuestra corrección.<br>
